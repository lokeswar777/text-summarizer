```python
import requests
from bs4 import BeautifulSoup
from transformers import pipeline

def scrape_wikipedia_body_text(url):
    """
    Scrapes the body text of a Wikipedia article given its URL.
    """
    response = requests.get(url)
    if response.status_code != 200:
        raise Exception(f"Failed to fetch the page. Status code: {response.status_code}")

    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the main content of the article
    paragraphs = soup.select("div.mw-parser-output > p")
    body_text = " ".join([para.text for para in paragraphs if para.text.strip()])

    return body_text.strip()

def summarize_text(text, chunk_size=1000, max_length=100, min_length=50):
    """
    Summarizes text in chunks to handle long inputs.
    """
    summarizer = pipeline("summarization", model="RussianNLP/FRED-T5-Summarizer")
    summaries = []

    for i in range(0, len(text), chunk_size):
        chunk = text[i:i + chunk_size]
        summary = summarizer(chunk, max_length=max_length, min_length=min_length, do_sample=False)
        summaries.append(summary[0]['summary_text'])

    return " ".join(summaries)

# Step 1: Provide the Wikipedia link
wiki_url = input("Enter the Wikipedia link: ")

try:
    # Step 2: Scrape the body text
    print("Scraping text from Wikipedia...")
    body_text = scrape_wikipedia_body_text(wiki_url)
    print(f"Scraped text length: {len(body_text)} characters")

    # Step 3: Summarize the text
    print("Summarizing text...")
    summarized_text = summarize_text(body_text)
    print("\n--- Summarized Text ---\n")
    print(summarized_text)

except Exception as e:
    print(f"An error occurred: {e}")
```
